```python
from bs4 import BeautifulSoup
import requests

# Step 5: If direct text extraction is not supported, download the filing content
response = requests.get(latest_filing.url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract specific sections
executives_section = soup.find('section', {'id': 'executives'})  # Hypothetical ID
board_members_section = soup.find('section', {'id': 'board_members'})  # Hypothetical ID

# Print extracted sections
print("Executives Section:")
print(executives_section.text[:1000])  # Print first 1000 characters

print("\nBoard Members Section:")
print(board_members_section.text[:1000])  # Print first 1000 characters
```
